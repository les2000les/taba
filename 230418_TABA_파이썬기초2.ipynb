##plot

```python
x = np.arange(0, 50, 0.1)
sin = np.sin(x)
cos = np.cos(x)
```

```python
plt.plot(x, sin, color = 'pink')
plt.plot(x, cos, color = 'skyblue')
plt.title('사인코사인 그래프')
plt.xlabel('x축')
plt.ylabel('y축')
plt.show()
```

```python
animals = ['사자', '얼룩말', '코끼리']
animals_in_A = [10, 5, 13]
animals_in_B = [3, 15, 3]

plt.bar(animals, animals_in_A, color = 'brown', width = 0.3, label = 'EOO동물원')
plt.bar(animals, animals_in_B, color = 'orange', width =0.3, label = 'SOO 동물원')
plt.title('동물원 안 동물들')
plt.xlabel('동물 이름')
plt.ylabel('동물 수 ')
plt.legend(loc = 'best')
plt.show()
```

```python
animals = ['사자', '얼룩말', '코끼리']
animals_in_A = [10, 5, 13]
animals_in_B = [3, 15, 3]

x = np.arange(1,4,1)
y = ['멸종위기', '적정', '포화']
w = 0.3

plt.bar([1,2,3], animals_in_A, color = 'brown', width = w, label = 'EOO동물원')
plt.bar([1+w, 2+w, 3+w], animals_in_B, color = 'orange', width = w, label = 'SOO 동물원')

plt.xticks(x+w/2, animals)
plt.yticks([2,5,10], y)
plt.title('동물원 안 동물들')
plt.xlabel('동물 이름')
plt.ylabel('동물 수 ')
plt.legend(loc = 'best')
plt.show()
```
